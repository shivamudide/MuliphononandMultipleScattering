```python
def func2(x, *params):
    y = np.zeros_like(x)
    for i in range(0, len(params), 3):
        ctr = params[i]
        amp = params[i+1]
        wid = params[i+2]
        y = y + amp * (wid / ((x - ctr)**2 + wid**2))
    return y


guess = [10, .06, 3, 12, .06, 3, 22, .06, 3, 28, .06, 3, 32, .06, 3, 35, .01, 3] 

popt, pcov = curve_fit(func2, X, S, p0=guess)
for i in range(0, len(popt), 3):
    print("Position: " + str(popt[i]) + "; Amplitude: " + str(popt[i+1]) + "; HWHM: " + str(popt[i+2]))
fit = func2(X, *popt)

plt.figure(figsize=(9, 9))
plt.plot(X, S, 'black')
plt.plot(X, fit , 'g-')
plt.title('Round-5 with Six Lorenzians', size=15)
plt.xlabel('Energy (meV)', size=15)
plt.ylabel('Intensity', size=15)
plt.show()
```

```python
from scipy.special import wofz
def func3(x, *params):
    y = np.zeros_like(x)
    for i in range(0, len(params), 4):
        ctr = params[i]
        amp = params[i+1]
        alpha = params[i+2]
        gamma = params[i+3]
        sigma = alpha / np.sqrt(2 * np.log(2))
        y = y + amp * np.real(wofz(((x-ctr) + 1j*gamma)/sigma/np.sqrt(2))) / sigma\
                                                           /np.sqrt(2*np.pi)
    return y

guess = [10, .06, 0.01, 2, 12, .06, 3, 2, 22, .06, 3, 2, 28, .06, 3, 2, 32, .06, 3, 2, 35, .01, 3, 2] 

popt, pcov = curve_fit(func3, X, S, p0=guess)
for i in range(0, len(popt), 4):
    print("Position: " + str(popt[i]) + "; Amplitude: " + str(popt[i+1]) + "; Alpha: " + str(popt[i+2]) + "; Gamma: " + str(popt[i+3]))
fit = func3(X, *popt)

plt.figure(figsize=(9, 9))
plt.plot(X, S, 'black')
plt.plot(X, fit , 'g-')
plt.title('Round-5 with Six Voigt Profiles', size=15)
plt.xlabel('Energy (meV)', size=15)
plt.ylabel('Intensity', size=15)
plt.show()
```
